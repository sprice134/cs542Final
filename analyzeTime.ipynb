```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('allBut17and21.csv').drop(columns=['Unnamed: 0'])
```

```python
# Calculating statistics
statistics = pd.DataFrame({
    'Mean': df.mean(),
    'Median': df.median(),
    'Minimum': df.min(),
    'Maximum': df.max(),
    'Standard Deviation': df.std(),
    'Std Dev / Mean': df.std() / df.mean()
})

# For the first plot, we drop 'Standard Deviation' from the columns for plotting
statistics_filtered = statistics.drop(columns=['Standard Deviation'])  # Correct usage here
statistics_filtered.plot(kind='bar', figsize=(14, 7), width=0.8)
plt.title('Mean, Median, Minimum, and Maximum of Each Query')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()

# For the second plot, we plot only 'Standard Deviation'
statistics[['Standard Deviation']].plot(kind='bar', figsize=(14, 7), color='c', width=0.8)
plt.title('Standard Deviation of Each Query')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()

statistics[['Std Dev / Mean']].plot(kind='bar', figsize=(14, 7), color='b', width=0.8)
plt.title('Standard Deviation Divided by Mean of Each Query')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```
